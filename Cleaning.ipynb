Cleaning Data Using Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Basic Data Exploration

```python
#Loading the Dataset
data = pd.read_csv("Crop Production data.csv")
```

```python
data.head()
```

```python
data.tail()
```

```python
#Checks for missing Values
data.isna().sum()
```

```python
#Dropping Rows with Null Values
data.dropna()
```

```python
data.describe()
```

Basic Data Visulization

```python
#Basic Data Visulization
plt.figure(figsize=(10,6))
sns.lineplot(data=data,x='Crop_Year',y='Production',size=20)
plt.xlabel('Year')
plt.ylabel('Production')
plt.title('Yearly Production')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=data,x='State_Name',y='Production')
plt.xlabel('State')
plt.ylabel('Production')
plt.title('Production by State')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=data,x='Season',y='Production')
plt.xlabel('Season')
plt.ylabel('Production')
plt.title('Production by Season')
plt.show()
```
